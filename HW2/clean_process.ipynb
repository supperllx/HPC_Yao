```python
def change_marked_variable2(mv, lower, upper):
    
    if lower<=mv<=upper:
        mark=1
    else:
        mv=0
        mark=0
    
    return mark

def change_marked_variable3(mv, lower, upper):
    
    if lower<=mv<=upper:
        mark_out=0
        mark_not=0
    else:
        if mv==000:
            mv=0
            mark_out=1
            mark_not=0
        else:
            mark_out=0
            mark_not=1
    
    return (mark_out, mark_not)

print('Done')
```

```python
def func(row):
    
    mark_fico = change_marked_variable2(row['fico'], 301, 805)
    mark_mi_pct_out, mark_mi_pct_not = change_marked_variable3(row['mi_pct'], 1, 55)
    mark_cltv = change_marked_variable2(row['cltv'], 0, 200)
    mark_dti = change_marked_variable2(row['dti'], 0, 65)
    mark_ltv = change_marked_variable2(row['ltv'], 6, 105)
    
    default = 1 if row['Default'] == 'YES' else 0
    
    return pd.Series([mark_fico, mark_mi_pct_out, mark_mi_pct_not, mark_cltv, mark_dti, mark_ltv, default])

data[['mark_fico', 'mark_mi_pct_out', 'mark_mi_pct_not', 'mark_cltv', 'mark_dti', 'mark_ltv', 'Default']] = data.swifter.apply(func, axis = 1)
data.head(20)
```

```python
data.cd_msa.fillna(value = 0, inplace=True)
data.ppmt_pnlty.fillna(value = 'Y', inplace=True)
data.zipcode.fillna(value = 0, inplace=True)
data.flag_sc.fillna(value = 'N', inplace=True)
data.cnt_units.values[data.cnt_units.values==99] = data.cnt_units.mode()[0]
```
